```python
#remove duplicate district-year rows
dup = df.groupby(["District", "Year"]).size().sort_values(ascending=False)

dup.head(20)
```

```python
df = df.groupby(["District", "Year"], as_index=False).agg({
    "Avg_Rainfall":"mean",
    "Area_Sown":"mean",
    "Wheat_Percentage":"mean",
    "Avg_Temperature":"mean",
    "Crop_Yield":"mean",
    "Irrigation_Area":"mean"
})
```

```python
df.head(15)
```

```python
df.shape
```

```python
#cleaning numeric cols
num_cols = ["Avg_Rainfall","Area_Sown","Wheat_Percentage","Avg_Temperature",
            "Crop_Yield","Irrigation_Area"]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
#handling missing vals

df.isna().sum()
```

```python
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

df.isna().sum()
```
