```python
import pandas as pd

#load area_sown.csv and convert wide format â†’ long format
area = pd.read_csv(r"C:\Ibrahim\Personal\University Stuff\Machine Learning\Project\ML Irrigation Project\data\raw\area_sown.csv")

area.head()
```

```python
area.rename(columns={"Province/division\n/district": "District"}, inplace=True)
```

```python
area.head()
```

```python
area.shape
```

```python
#melt into long format (wide year col -> rows)
area_long = area.melt(
    id_vars = ["District"],
    var_name = "YearRange",
    value_name = "Area_Sown"
)

#extract first year from "2007-08"
area_long["Year"] = area_long["YearRange"].str.split("-").str[0].astype(int)

#drop extra
area_long = area_long[["District", "Year", "Area_Sown"]]
```

```python
area_long.head()
```

```python
area_long.shape
```

```python
#load wheat_pct.csv - transform and merge
wheat = pd.read_csv(r"C:\Ibrahim\Personal\University Stuff\Machine Learning\Project\ML Irrigation Project\data\raw\wheat_pct.csv")

wheat.shape
```

```python
wheat.head()
```

```python
wheat.rename(columns={"Province/division\n/district": "District"}, inplace=True)
```

```python
wheat.head()
```

```python
wheat_long = wheat.melt(
    id_vars = ["District"],
    var_name = "YearRange",
    value_name = "Wheat_Percentage"
)
```

```python
wheat_long["Year"] = wheat_long["YearRange"].str.split("-").str[0].astype(int)

wheat_long = wheat_long[["District", "Year", "Wheat_Percentage"]]
```

```python
wheat_long.head()
```

```python
wheat_long.shape
```

```python
#merge with base table
df = area_long.merge(wheat_long, on=["District", "Year"], how="left")

df.head(10)
```
