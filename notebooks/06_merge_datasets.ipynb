```python
import pandas as pd
import numpy as np

rain = pd.read_csv(
    r"C:\Ibrahim\Personal\University Stuff\Machine Learning\Project\ML Irrigation Project\data\raw\Rain_fall_in_Pakistan.csv"
)

#remove metadata row
rain = rain.iloc[1:].copy()

#parse date
rain["date"] = pd.to_datetime(rain["date"], errors="coerce")

#convert main rainfall columns to numeric
rain_cols = ["rfh", "rfh_avg", "r1h", "r1h_avg", "r3h", "r3h_avg"]
rain[rain_cols] = rain[rain_cols].apply(pd.to_numeric, errors="coerce")

#extract Year for grouping
rain["Year"] = rain["date"].dt.year

#create YEARLY rainfall summary
rain_yearly = (
    rain.groupby("Year")[rain_cols].mean().reset_index()
)

rain_yearly.rename(columns={"rfh": "Avg_Rainfall"}, inplace=True)

rain_yearly.head()
```

```python
rain_yearly.shape
```

```python
#(temp_pk.csv)
temp = pd.read_csv(
    r"C:\Ibrahim\Personal\University Stuff\Machine Learning\Project\ML Irrigation Project\data\raw\temp_pk.csv"
)

#convert month+year â†’ datetime
temp["date"] = pd.to_datetime(
    temp["Year"].astype(str) + "-" + temp["Month"] + "-01",
    errors="coerce"
)

#convert temperature column to numeric
temp["Temperature"] = pd.to_numeric(temp["Temperature - (Celsius)"], errors="coerce")

#extract year
temp["Year"] = temp["date"].dt.year

#compute annual temperature average
temp_yearly = temp.groupby("Year")["Temperature"].mean().reset_index()
temp_yearly.rename(columns={"Temperature": "Avg_Temperature"}, inplace=True)

temp_yearly.head()
```

```python
temp_yearly.shape
```

```python
area = pd.read_csv(
    r"C:\Ibrahim\Personal\University Stuff\Machine Learning\Project\ML Irrigation Project\data\raw\area_sown.csv"
)

area.rename(columns={"Province/division\n/district": "District"}, inplace=True)

#melt wide data into long format
area_long = area.melt(
    id_vars="District",
    var_name="Year",
    value_name="Area_Sown"
)

#convert Year like "2007-08" â†’ 2007
area_long["Year"] = area_long["Year"].str.split("-").str[0].astype(int)

#clean numeric formatting
area_long["Area_Sown"] = (
    area_long["Area_Sown"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .replace({"**": np.nan, "-": np.nan, "": np.nan})   # <- FIX
)

#some entries may still contain whitespace or weird characters so we force numeric safely
area_long["Area_Sown"] = pd.to_numeric(area_long["Area_Sown"], errors="coerce")

area_long.head()
```

```python
import pandas as pd
import numpy as np

rain = pd.read_csv(r"C:\Ibrahim\Personal\University Stuff\Machine Learning\Project\ML Irrigation Project\data\raw\Rain_fall_in_Pakistan.csv")

# remove meta row if present
if rain.iloc[0,0].startswith("#"):
    rain = rain.iloc[1:].copy()

rain['date'] = pd.to_datetime(rain['date'], errors='coerce')
rain['Year'] = rain['date'].dt.year

rain['rfh'] = pd.to_numeric(rain['rfh'], errors='coerce')

# compute annual district rainfall
rain_yearly = (
    rain.groupby(["ADM2_PCODE", "Year"])["rfh"]
        .mean()
        .reset_index()
        .rename(columns={"ADM2_PCODE": "District", "rfh": "Avg_Rainfall"})
)

rain_yearly.head()
```
